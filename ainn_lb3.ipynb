```python
from PIL import Image
import numpy as np
import keras
import matplotlib.pyplot as plt
import os
```

```python
def download_data(path):
  data = []
  for path_image in sorted(os.listdir(path=path)):
    full_path = path + '/' + path_image
    image = Image.open(full_path)
    w, h = image.size
    if (w != 512): # test label имеют размер (256, 256, 4)
      image = image.resize((512, 512)) # изменение размера
      image = image.convert('1') # приведение к 1 каналу
    data.append(np.array(image))
  return data

X_train = download_data(r"ISBI-2012/train/image")
Y_train = download_data(r"ISBI-2012/train/label")
X_test = download_data(r"ISBI-2012/test/image")
Y_test = download_data(r"ISBI-2012/test/label")
```

```python
plt.figure(figsize=(12, 12))
for i in range(8):
  plt.subplot(4, 4, i+1)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(X_train[i], cmap=plt.cm.binary)
for i in range(8):
  plt.subplot(4, 4, i+1+8)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(Y_train[i], cmap=plt.cm.binary)
```

```python
# 30 изображений 512 на 512 в градациях серого (1 канал)
X_train_prep = np.array(X_train).reshape([30, 512, 512, 1]) / 255
Y_train_prep = np.array(Y_train).reshape([30, 512, 512, 1]) / 255

print(Y_test[0].shape)
# 30 изображений 512 на 512 в градациях серого (1 канал)
X_test_prep = np.array(X_test).reshape([30, 512, 512, 1]) / 255
Y_test_prep = np.array(Y_test).reshape([30, 512, 512, 1]) / 255


# так как тестовая и тренировочная выборка равны по размеру
# решено для обучения объединить половину тестовой выборки
# и изначальную тренировочную
X_test_new, X_test_to_merge = np.split(X_test_prep, 2, axis=0)
Y_test_new, Y_test_to_merge = np.split(Y_test_prep, 2, axis=0)

print(X_test_to_merge.shape)
print(X_train_prep.shape)

X_train_new = np.concatenate([X_test_to_merge, X_train_prep], axis=0)
Y_train_new = np.concatenate([Y_test_to_merge, Y_train_prep], axis=0)

print(Y_train_new.shape)
```

```python
img_path = 'mangadood.png'
img = Image.open(img_path)
img = img.convert('1')
img = img.resize((512, 512))
input = np.array(img)
input = np.array(input).reshape([1, 512, 512, 1]) / 255
```

```python
def double_conv(input, f1, f2):
    out = keras.layers.Conv2D(filters=f1,
                                kernel_size=3,
                                padding='same',
                                strides=(1, 1),
                                activation='relu')(input)
    out = keras.layers.Conv2D(filters=f2,
                                kernel_size=3,
                                padding='same',
                                strides=(1, 1),
                                activation='relu')(out)
    return out
```

```python
input = keras.layers.Input(shape=(512, 512, 1))
# encoder
conv_1 = double_conv(input, 16, 16)
maxp_1 = keras.layers.MaxPooling2D(pool_size=4, strides=(4, 4))(conv_1)

conv_2 = double_conv(maxp_1, 16, 16)
maxp_2 = keras.layers.MaxPooling2D(pool_size=2, strides=(2, 2))(conv_2)

bottleneck = maxp_2

#decoder
up_2 = keras.layers.Conv2DTranspose(filters=32,
                                    kernel_size=2,
                                    strides=(2, 2),
                                    padding='same')(bottleneck)
conc_2 = keras.layers.concatenate([up_2, conv_2])

up_1 = keras.layers.Conv2DTranspose(filters=32,
                                    kernel_size=4,
                                    strides=(4, 4),
                                    padding='same')(conc_2)
conc_1 = keras.layers.concatenate([up_1, conv_1])

output = keras.layers.Conv2D(filters=1,
                             kernel_size=1,
                             padding='same',
                             strides=(1, 1),
                             activation='sigmoid')(conc_1 * 8)


model = keras.Model(input, output)
```

```python
# компиляция НС
adam = keras.optimizers.Adam(learning_rate=0.01)
model.compile(loss='binary_crossentropy', metrics=['binary_accuracy'], optimizer=adam)
```

```python
log = model.fit(X_train_new, Y_train_new, epochs=700, validation_split=0.2)
```
